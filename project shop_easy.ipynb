```python
# Customer Reviews Analysis
# Highest rated:

high_rating = """SELECT p.ProductID, p.ProductName, cr.Highest_Rating
    FROM shop.products AS p
    INNER JOIN (
        SELECT ProductID, MAX(Rating) AS Highest_Rating
        FROM shop.customer_reviews
        GROUP BY ProductID
        having max(rating)>3
    ) AS cr ON p.ProductID = cr.ProductID;
"""

curr.execute(high_rating)
columns=curr.column_names
results=curr.fetchall()
data_frame=pd.DataFrame(results,columns=columns)
data_frame
```

```python
# sentiment analysis:
feedback=" select CustomerID, ReviewText from customer_reviews"
curr.execute(feedback)
columns=curr.column_names
feedback_data = curr.fetchall()
data_frame=pd.DataFrame(feedback_data,columns=columns)
data_frame
```

```python
# Analyse sentiment analysis of the customer

from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
analyzer = SentimentIntensityAnalyzer()


def analyze_sentiment(feedback):
    sentiment_scores = analyzer.polarity_scores(feedback)
    compound_score = sentiment_scores['compound']
    
    # Determine sentiment as positive, negative, or neutral
    if compound_score >= 0.05:
        return 'Positive'
    elif compound_score <= -0.05:
        return 'Negative'
    else:
        return 'Neutral'

# Process each feedback entry
for feedback_entry in feedback_data:
    customer_id = feedback_entry[0]
    feedback = str(feedback_entry[1])  
    sentiment = analyze_sentiment(feedback)
    print(f"Customer ID: {customer_id}, Sentiment: {sentiment}")
```

```python
# Add sentiment column to existing column

sentiment=""" ALTER TABLE customer_reviews ADD COLUMN feedback VARCHAR(10);
"""
curr.execute(sentiment)
db_connection.commit()
```

```python
# updating sentiment data 


for feedback_entry in feedback_data:
    customer_id = feedback_entry[0]
    feedback = feedback_entry[1]
    sentiment = analyze_sentiment(feedback)
    
    update_query = """
        UPDATE customer_reviews
        SET feedback = %s
        WHERE CustomerID = %s
    """
    curr.execute(update_query, (sentiment, customer_id))
    db_connection.commit()
```

```python
#Marketing Effectiveness:
#customer retention rate:

start_date ='2023-01-01'
end_date='2025-12-31'
customer_start = """
SELECT COUNT(DISTINCT CustomerID)
FROM customer_journey
WHERE VisitDate <= %s
"""


customer_end = """
SELECT COUNT(DISTINCT CustomerID)
FROM customer_journey
WHERE VisitDate <= %s
"""


new_customers = """
SELECT COUNT(DISTINCT CustomerID)
FROM customer_journey
WHERE VisitDate BETWEEN %s AND %s
"""

curr.execute(customer_start, (start_date,))
customers_visit = curr.fetchone()[0]

curr.execute(customer_end, (end_date,))
customers_quit = curr.fetchone()[0]

curr.execute(new_customers, (start_date, end_date))
new_visit = curr.fetchone()[0]


retained_customers = customers_quit - new_visit


if customers_visit > 0:
    retention_rate = (retained_customers / customers_visit) * 100
    print(f"Customer Retention Rate: {retention_rate:.2f}%")
else:
    print("No customers repeated")
```

```python
# compare first time buyers vs repeated buyers:


start_date ='2023-01-01'
end_date='2025-12-31'
customer_start = """
SELECT COUNT(DISTINCT CustomerID)
FROM customer_journey
WHERE VisitDate <= %s
"""


customer_end = """
SELECT COUNT(DISTINCT CustomerID)
FROM customer_journey
WHERE VisitDate <= %s
"""


first_time = """
    SELECT CustomerID, MIN(VisitDate) AS FirstPurchaseDate
    FROM customer_journey
    GROUP BY CustomerID
    HAVING FirstPurchaseDate BETWEEN %s AND %s;
"""


repeat = """
    SELECT CustomerID, COUNT(JourneyID) AS PurchaseCount
    FROM customer_journey
    GROUP BY CustomerID
    HAVING PurchaseCount > 1 AND MIN(VisitDate) BETWEEN %s AND %s;
"""


curr.execute(first_time, (start_date, end_date))
first_time_buyers = curr.fetchall()


curr.execute(repeat, (start_date, end_date))
repeat_buyers = curr.fetchall()


first_time_buyer_ids = [row[0] for row in first_time_buyers]
repeat_buyer_ids = [row[0] for row in repeat_buyers]

print(f"First-time Buyers: {len(first_time_buyer_ids)}")
print(f"Repeat Buyers: {len(repeat_buyer_ids)}")
```

```python
#best-performing products per region:

best_performing_products=""" select g.Country,p.ProductName,COUNT(Action) as Best_products
                             from products as p
                             INNER JOIN customer_journey as cj ON p.ProductID=cj.ProductID
                             INNER JOIN customers as c
                             ON(cj.CustomerID=c.CustomerID)
                             INNER JOIN geography as g
                             ON(c.GeographyID=g.GeographyID)
                             WHERE cj.Action = 'purchase'
                            GROUP BY g.Country, p.ProductName
                            ORDER BY Best_products DESC;
"""
curr.execute(best_performing_products)
columns=curr.column_names
results=curr.fetchall()
data_frame=pd.DataFrame(results,columns=columns)
data_frame
```
